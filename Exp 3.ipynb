```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
# Reading the dataset and adding the realted headers
filename = "auto.csv"
```

```python
headers= ["symboling","normalized-losses","make","fuel-type","aspiration","num-of-doors","body-style","drive-wheels",
          "engine-location","wheel-base","length","width","height","curb-height","engine-type","num-of-cylinders",
          "engine-size","fuel-systems","bore","stroke","compression-ratio","horsepower","peak-rpm","city-mpg",
          "highway-mpg","price"]
```

```python
# use the Pandas method read_csv() to load the data. Set the parameters "names" equal to the Python list "Headers"
df = pd.read_csv(filename, names= headers)
```

```python
# to see what the data set looks like, we'll use the head() method.
df.head()
```

```python
# Import and handle missing values
# Convert "?" to NaN

import numpy as np
# replace "?" to NaN
df.replace("?", np.nan,inplace=True)
df.head(5)
```

```python
missing_data = df.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

```python
# Caluculate the mean value for the "normalized-losses" column
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
# Replace "NaN" with mean value in "nprmalized-losses"column
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
#Calulate the mean value for the "bore" column
avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
```

```python
#Replace "Nan" with the mean value in the "bore" column
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
#Write your code below and press Shift+Enter to execute
avg_stroke=df['stroke'].astype('float').mean(axis=0)
print("Average of stroke:", avg_stroke)
```

```python
df["stroke"].replace(np.nan, avg_bore, inplace=True)
```

```python
# Calculate the mean value for the "horsepower"column
avg_horsepower=df['horsepower'].astype('float').mean(axis=0)
print("Average of horsepower:", avg_horsepower)
```

```python
df["horsepower"].replace(np.nan, avg_bore, inplace=True)
```

```python
#Calculate the mean value for the "peak-rpm" column
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average of peak-rpm:", avg_peakrpm)
```

```python
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
#replace the missing 'num-of-doors' values by the most frequent
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
#simply drop while row with NaN in "price" column
df.dropna(subset=["price"],axis=0, inplace=True)

#reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
```
